```python
from scipy.integrate import odeint

def dP_dt (P,t):
    dP=0.03067*P
    return dP


año= np.arange(1790,2000,10)
t = np.arange(0,210,10)
real = np.array([3.9, 5.3, 7.2, 9.6, 12, 17, 23, 31, 38, 50, 62, 75, 91, 105, 122, 131, 151, 179, 203, 226, 249]) #linea azul

Pt = 3.9*np.exp(0.03067*t)

plt.plot(t,real,label="Real")
plt.plot(t,Pt,label="Analiticamente")

#t=np.arange(0,3,0.1)

x = odeint(dP_dt,3.9,t)
plt.plot(t,x,label="odeint")


_=plt.xticks(t, año, rotation='vertical')
plt.margins(0.05)
plt.xlabel("t")
plt.ylabel("P(t)")
plt.legend(loc=10)
```

```python
from scipy.integrate import odeint
N=210
def dP_1 (P,t):
    dP_1=0.03067*(1-P/N)*P
    return dP_1

t1 = np.arange(0,210,10)
x1 = odeint(dP_1,3.9,t1)


plt.plot(t1,x1)
plt.plot(t,real,label="Real")
```
