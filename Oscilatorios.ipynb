$$\begin{eqnarray}
\dot{x} &=&  y \\
\dot{y} &=& -\frac{k}{m} x - \frac{B}{m} y
\end{eqnarray}$$

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import odeint
%matplotlib inline
```

```python
k = 3.0
m = 1.0 
B = 0.5

def armonico(variables, t):
    x, y = variables
    return [y, -k* x / m - B / m * y]

inicial = [0.8, 0]

tiempo = np.arange(0.0, 20.0, 0.01)

resultado = odeint(armonico, inicial, tiempo)

xx, yy = resultado.T

plt.plot(tiempo, xx, c = 'r', label="Posicion")
plt.plot(tiempo, yy, c = 'k', label="Velocidad")
plt.legend(loc = 'best', prop ={'size':14})
plt.xlabel('tiempo', fontsize=14)
plt.show()
```

```python
from ipywidgets import *
```

```python
def amortiguador(t=0):
    fig = plt.figure()
    ax = fig.add_subplot(1,1,1)
    ax.plot(tiempo, xx, '-k')
    ax.plot(tiempo[t], xx[t], 'ro')
    ax.text(4, 0.6, 'tiempo =  %s'%tiempo[t])
    ax.text(4, 0.5, 'posicion =  %s'%xx[t])
    fig.canvas.draw()
    
t_f = len(tiempo)
interact_manual(amortiguador, t= (0, t_f, 1))
```
